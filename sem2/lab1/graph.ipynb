```python
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats

set_matplotlib_formats ('png')
```

```python
plt.figure(figsize=(16,16))
ax1 = plt.axes()

x = [1, 2, 4]
y = [0.015482, 0.008564, 0.004906]
plt.plot (x, y, label="Задача 1Д")

x = [i for i in range (1, 13)]
y = [0.013092, 0.010989, 0.008518, 0.006867, 0.006952, 0.006612,
     0.005663, 0.006165, 0.005807, 0.005583, 0.006379, 0.006484]
plt.plot (x, y, label="Задача 2Г")

x = [1, 2, 3, 4, 5, 6]
y = [0.014494, 0.007650, 0.005237, 0.004198, 0.005368, 0.004227]
plt.plot (x, y, label="Задача 3В")

ax1.set_xlabel ("Количество потоков")
ax1.set_ylabel ("Время работы (среднее для потока) (в мс)")

plt.grid()
plt.legend()

plt.savefig("./picturies/graph.png", dpi=100)
plt.show()
```
