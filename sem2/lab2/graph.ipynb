```python
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats

set_matplotlib_formats ('png')

import pandas as pd
import numpy as np
```

```python
plt.figure(figsize=(16,16))
ax1 = plt.axes()

with open('result.txt') as f:
	lines = f.readlines()

	for line in lines:
		numbers = list(map(float, line.split()))
		b = numbers[0]
		y = numbers[1:]
		count = len(y)
		x = [i * 1.0 / count for i in range(count)]

		plt.plot (x, y, label=f"b = {b}")

plt.grid()
plt.legend()

plt.savefig("./picturies/graph.png", dpi=100)
plt.show()
```
